```python
import openai
import gradio as gr

openai.api_key = 'sk-proj-x54ytxmoyfoidfe7wztlty6j8ssrkgvsk76iynainf95c51s2jj3tjjcghlnkfxkyl598isrjlt3blbkfjllynmxsxpsnenyrx0ptnofnt4p088b9omejuctynpmnskg8vpbfatg7-avu-koy9cmlwasoica'

def chat_with_gpt(prompt):
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "system", "content": "You are a helpful assistant."},
            {"role": "user", "content": prompt}
        ]
    )
    reply = response['choices'][0]['message']['content']
    return reply

def chatbot_interface(user_input):
    bot_response = chat_with_gpt(user_input)
    return bot_response

iface = gr.Interface(
    fn=chatbot_interface,
    inputs="text",
    outputs="text",
    title="Varma's Chatbot",
    description="Talk to a ChatGPT-based assistant.",
)

iface.launch()
```
